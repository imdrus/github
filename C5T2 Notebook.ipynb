```python
import numpy as np
```

```python
# Balance limit of all customers by marriage and education . Also total was found for each class grouped by sex.
#  marriage (1 = married; 2 = single; 3 = divorce; 0=others)
#  education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others)
#  Largest Total balances are held by Singles with grad education or married with University eduction
credit.pivot_table('LIMIT_BAL', 'MARRIAGE', 'EDUCATION', aggfunc=np.sum, margins=True)
```

```python
# Default next month of all customers by marriage and education . Also total was found for each class grouped by sex.
#  marriage (1 = married; 2 = single; 3 = divorce; 0=others)
#  education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others)
#  Largest Total balances are held by Singles with grad education or married with University eduction
credit.pivot_table('DefaultPaymentNextMonth', 'MARRIAGE', 'EDUCATION', aggfunc=np.sum, margins=True)
```

```python
# Default next month of all customers by AGE and education . Also total was found for each class grouped by sex.
# bins=[20,25,35,50,65,80]
#  marriage (1 = married; 2 = single; 3 = divorce; 0=others)
#  education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others)
#  Most defaults are in the 25-35 age category
credit.pivot_table('DefaultPaymentNextMonth', 'AGE', 'EDUCATION', aggfunc=np.sum, margins=True)
```

```python
# Passengers who survived and who didn't survive grouped by class and sex
# Y=0 then not default, Y=1 then default
# Education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others)
# Age bins=[20,25,35,50,65,80]
table = pd.crosstab(index=[credit.AGE,credit.SEX], columns=[credit.DefaultPaymentNextMonth,credit.EDUCATION])
```

```python
table.unstack()
```

```python
# Passengers who survived and who didn't survive grouped by class and sex
# Y=0 then not default, Y=1 then default
# Education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others)
# Age bins=[20,25,35,50,65,80]
# (1 = married; 2 = single; 3 = divorce; 0=others)
table = pd.crosstab(index=[credit.AGE,credit.MARRIAGE], columns=[credit.DefaultPaymentNextMonth,credit.EDUCATION])
```
